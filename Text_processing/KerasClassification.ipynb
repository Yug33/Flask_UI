```python
import pandas as pd
import sys  
import numpy as np
from sklearn.model_selection import train_test_split
from keras.preprocessing import text, sequence
from sklearn.preprocessing import LabelBinarizer, LabelEncoder
from keras.layers import Dense, Activation, Dropout
```

```python
filepath = 'input/data_set_for_NN.csv'
df = pd.read_csv(filepath, encoding='latin-1')

df['CATEGORY'] = df.CATEGORY.map({'b': 0, 't': 1, 'e': 2, 'm': 3})
sentences = df['TITLE']
y = df['CATEGORY']

sentences_train, sentences_test, y_train, y_test = train_test_split(    
    sentences, y, test_size=0.25, random_state=1000)
```

```python
# from imp import reload
# 
# reload(sys)  
# sys.setdefaultencoding('utf-8')
# 
# s = pd.concat([sentences_test, y_test], axis=1)
# s.to_csv('sample_data/sample_test.csv', sep=',')
```

```python
max_words = 1000
tokenize = text.Tokenizer(num_words=max_words, char_level=False)
```

```python
tokenize.fit_on_texts(sentences_train) 
x_train = tokenize.texts_to_matrix(sentences_train)
x_test = tokenize.texts_to_matrix(sentences_test)
```

```python
encoder = LabelEncoder()
encoder.fit(y_train)
y_train = encoder.transform(y_train)
y_test = encoder.transform(y_test)
```

```python
from keras import utils

num_classes = np.max(y_train)+1
y_train = utils.to_categorical(y_train, num_classes)
y_test = utils.to_categorical(y_test, num_classes)
```

```python
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
batch_size = 32
epochs = 2
```

```python
# Build the model
from keras import Sequential

model = Sequential()
model.add(Dense(512, input_shape=(max_words,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_split=0.1)
```

```python
my_y = model.predict_classes(x_test)
np.savetxt("test_output.csv", my_y, delimiter=",")
```

```python
score = model.evaluate(x_test, y_test,
                       batch_size=batch_size, verbose=1)
print('Test accuracy:', score[1])
```
